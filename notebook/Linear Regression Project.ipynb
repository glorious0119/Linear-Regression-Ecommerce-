## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
y = customers['Yearly Amount Spent']
```

```python
X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
# Function to train the model
def fit_model(X_train, y_train, model_type='linear', alpha=1.0, l1_ratio=0.5):
    """
    Instantiates and fits a regression model based on the specified type.

    Parameters:
        X_train (array-like): Training features.
        y_train (array-like): Training target variable.
        model_type (str): The type of model to instantiate. Options are 'linear', 'lasso', 'ridge', 'elasticnet'.
        alpha (float): Regularization strength; applies to Lasso, Ridge, and ElasticNet models.
        l1_ratio (float): Only used for ElasticNet; the mixing ratio between Lasso (L1) and Ridge (L2) penalties.

    Returns:
        model: The fitted model.
    """
    # Instantiate the model based on the model type
    if model_type == 'linear':
        model = LinearRegression()
    elif model_type == 'lasso':
        model = Lasso(alpha=alpha)
    elif model_type == 'ridge':
        model = Ridge(alpha=alpha)
    elif model_type == 'elasticnet':
        model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
    else:
        raise ValueError("Invalid model type. Choose from 'linear', 'lasso', 'ridge', 'elasticnet'.")

    # Fit the model
    model.fit(X_train, y_train)
    
    return model
```

```python
# Linear Regression (default model type)
linear_model = fit_model(X_train_scaled, y_train)
print("Linear Regression Coefficients:", linear_model.coef_)
print("Linear Regression Intercept:", linear_model.intercept_)
```

```python
# Lasso Regression
lasso_model = fit_model(X_train_scaled, y_train, model_type='lasso', alpha=0.1)
print("Lasso Regression Coefficients:", lasso_model.coef_)
print("Lasso Regression Intercept:", lasso_model.intercept_)
```

```python
# Ridge Regression
ridge_model = fit_model(X_train_scaled, y_train, model_type='ridge', alpha=1.0)
print("Ridge Regression Coefficients:", ridge_model.coef_)
print("Ridge Regression Intercept:", ridge_model.intercept_)
```

```python
# ElasticNet Regression
elasticnet_model = fit_model(X_train_scaled, y_train, model_type='elasticnet', alpha=0.1, l1_ratio=0.5)
print("ElasticNet Regression Coefficients:", elasticnet_model.coef_)
print("ElasticNet Regression Intercept:", elasticnet_model.intercept_)
```

## Predicting Test Data

```python
from sklearn import metrics
```

```python
def predict_model(model, X_test):
    """
    Uses a trained model to predict on the test dataset.

    Parameters:
        model: Trained model (e.g., LinearRegression, Lasso, Ridge, ElasticNet).
        X_test (array-like): Test features.

    Returns:
        array-like: Predictions for the test set.
    """
    predictions = model.predict(X_test)
    return predictions
```

```python
# Linear prediction
linear_predictions = predict_model(linear_model, X_test_scaled)
```

```python
# Lasso prediction
lasso_predictions = predict_model(lasso_model, X_test_scaled)
```

```python
# Ridge prediction
ridge_predictions = predict_model(ridge_model, X_test_scaled)
```

```python
# ElasticNet prediction
elasticnet_predictions = predict_model(elasticnet_model, X_test_scaled)
```

## Plot Actual vs. Predicted Values

```python
def plot_predictions(y_test, predictions):
    """
    Plots a scatter plot comparing actual vs. predicted values.

    Parameters:
        y_test (array-like): True target values from the test set.
        predictions (array-like): Predicted values for the test set.
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, predictions, color='blue', edgecolor='k', alpha=0.6)
    plt.xlabel('Actual Y Values')
    plt.ylabel('Predicted Y Values')
    plt.title('Actual vs Predicted Values')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
    plt.show()
```

```python
plot_predictions(y_test, linear_predictions)
```

```python
plot_predictions(y_test, lasso_predictions)
```

```python
plot_predictions(y_test, ridge_predictions)
```

```python
plot_predictions(y_test, elasticnet_predictions)
```

## Evaluating the Model Performance

```python
def evaluate_model(y_test, predictions):
    """
    Evaluates model performance using MAE, MSE, RMSE, R², and adjusted R².

    Parameters:
        y_test (array-like): True target values from the test set.
        predictions (array-like): Predicted values for the test set.
        X_test (array-like, optional): Test features, required for adjusted R² calculation.

    Returns:
        dict: A dictionary with calculated metrics (MAE, MSE, RMSE, R², adjusted R²).
    """
    # Calculate MAE, MSE, and RMSE
    mae = metrics.mean_absolute_error(y_test, predictions)
    mse = metrics.mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    
    # Calculate R² Score
    r2 = metrics.r2_score(y_test, predictions)
    
    # Calculate Adjusted R² Score (only if X_test is provided)
    if X_test is not None:
        n = len(y_test)       # Number of samples
        p = X_test.shape[1]   # Number of features
        adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    else:
        adjusted_r2 = None  # If X_test is not provided, adjusted R² is unavailable
    
    # Print metrics
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R²:', r2)
    if adjusted_r2 is not None:
        print('Adjusted R²:', adjusted_r2)
    
    # Return metrics as a dictionary
    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'Adjusted R2': adjusted_r2
    }
```

```python
linear_evaluation_metrics = evaluate_model(y_test, linear_predictions)
```

```python
lasso_evaluation_metrics = evaluate_model(y_test, lasso_predictions)
```

```python
ridge_evaluation_metrics = evaluate_model(y_test, ridge_predictions)
```

```python
elasticnet_evaluation_metrics = evaluate_model(y_test, elasticnet_predictions)
```

```python
pd.DataFrame([linear_evaluation_metrics, lasso_evaluation_metrics, ridge_evaluation_metrics, elasticnet_evaluation_metrics],
            index=['Linear Regression', 'Lasso Regression', 'Ridge Regression', 'ElasticNet Regression'])
```

### Metrics Analysis
- Linear Regression has the lowest MAE, MSE, and RMSE, indicating smaller errors on average compared to the other models.
- R² and Adjusted R² values are also highest for Linear Regression, suggesting it explains the variance in the target variable better than the others

Based on this comparison, **Linear Regression performs slightly better across all metrics**, making it the best choice among these models.
